```python
import numpy as np
import numpy.linalg as lin
import cv2
import matplotlib.pyplot as mt
```

```python
def plot(img,out):
    mt.subplot(121)
    mt.title('Original image')
    mt.imshow(img)

    mt.subplot(122)
    mt.title('Tranformed image')
    mt.imshow(out)
    
    mt.show()
```

```python
U=[[10, 15, 1],[8, 3, 1],[11, 17, 1],[5, 11, 1],[6, 13, 1]]
X=[[33, 20, 1],[18, 7, 1],[37, 22, 1],[20, 13, 1],[23, 16, 1]]
U , X = np.array(U) , np.array(X)
T=np.round(lin.inv(U.T@U)@U.T@X, 4 )
T
```

```python
img=cv2.imread('cameramanimage_22843.tif')
mt.imshow(img)
```

```python
rows,cols,channels= img.shape
out = cv2.warpAffine(img,T[0:2],(cols,rows))
plot(img,out)
mt.imsave("Transformed img.jpg",out)
```

```python
img = cv2.imread('Transformed img.jpg')
row,col,ch = img.shape
out = cv2.warpAffine(img,lin.inv(T)[0:2,],(col,row))
plot(img,out)
mt.imsave("Inverse Transformed img.jpg",out)
```
